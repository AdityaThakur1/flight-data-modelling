```python
from pyspark.sql import SparkSession
MAX_MEMORY = "45g"


spark = SparkSession \
    .builder \
    .appName("airline-delay-1") \
    .config("spark.executor.memory", MAX_MEMORY) \
    .config("spark.driver.memory", MAX_MEMORY) \
    .getOrCreate()
```

```python
dfx = spark.read.options(header='True', inferSchema='True',mode='DROPMALFORMED').csv('drive/MyDrive/data/clean/2018.csv')
dfx = dfx.fillna({'DEP_DELAY': '0.0'})
dfx = dfx.fillna({'ARR_DELAY': '0.0'})
```

```python
dfx.show()
```

# train / test split

```python
features = ['DEP_DELAY','ORIGIN_INDEXED', 'OP_CARRIER_INDEXED', 'DEST_INDEXED']
target = 'ARR_DELAY'
dfx = dfx.na.drop()
data = dfx.select(features+[target])
data.show()
```

```python
X = data.toPandas()[features].values
```

```python
y = data.toPandas()['ARR_DELAY'].values
```

```python
import xgboost as xgb
#data_dmatrix = xgb.DMatrix(data=X, label=y)
```

```python
from sklearn.model_selection import train_test_split
#X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123)
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn import linear_model

#xg_reg = xgb.XGBRegressor(n_estimators=100, max_depth=7 ,seed = 123)

model_r = make_pipeline(StandardScaler(with_mean=True), xgb.XGBRegressor(n_estimators=100, max_depth=7))
```

```python
#xg_reg.fit(X, y)
model_r.fit(X, y.ravel())
```

```python
#preds = xg_reg.predict(X)
preds = model_r.predict(X)
```

```python
#cross validation

data_dmatrix = xgb.DMatrix(data=X,label=y)
```

```python
params = {"objective":"reg:linear",'colsample_bytree': 0.3,'learning_rate': 0.1, 'max_depth': 7, 'alpha': 10,  n_estimators = 1000}

cv_results = xgb.cv(dtrain=data_dmatrix, params=params, nfold=3, num_boost_round=50,early_stopping_rounds=10,metrics="rmse", as_pandas=True, seed=123)
```
